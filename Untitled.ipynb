```python
import nltk as nl
import random
```

```python
from nltk.corpus import names
```

```python
names.fileids()
```

```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use("ggplot")
%matplotlib inline
```

```python
name_cfd = nl.ConditionalFreqDist((fileid, name[-2:]) for fileid in names.fileids() for name in names.words(fileid))
```

```python
plt.figure(figsize = (40, 10))
name_cfd.plot()
```

```python
def name_features(name):
    return{"pair": name[-2:]}
```

```python
name_features("Katy")
```

```python
name_list = [(name, "male") for name in names.words('male.txt')] + [(name, "female") for name in names.words('female.txt')]
```

```python
name_list
```

```python
name_list_randomize = random.shuffle(name_list)
```

```python
features = [(name_features(name), gender) for (name, gender) in name_list]
```

```python
len(features) / 2
```

```python
training_set = features[:int((len(features) / 2))]
testing_set = features[int((len(features)/ 2)):]
```

```python
classifier = nl.NaiveBayesClassifier.train(training_set)
```
